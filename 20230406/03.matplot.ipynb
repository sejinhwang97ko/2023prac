```python
import seaborn as sns
```

```python
import pandas as pd
drinks = pd.read_csv('../csv/csv/drinks.csv')
drinks
```

```python
corr = drinks[
    [
        'beer_servings',
        'spirit_servings',
        'wine_servings',
        'total_litres_of_pure_alcohol'
    ]
    ].corr(method='pearson')

corr
```

```python
# 히트맵 출력
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

sns.set(font_scale=1.5) #글자크기 늘려주기

hm = sns.heatmap(
    corr.values,
    cbar= True,
    annot = True, # 히트맵에 숫자 표시
    square = True, # 정사각형
    fmt = '.2f',
    annot_kws= {'size': 20}, #글자 크기
    xticklabels= cols_view, #x축 이름
    yticklabels = cols_view
)

plt.tight_layout()
plt.show()
```

```python
# 히트맵 출력
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

sns.set(font_scale=1.5) #글자크기 늘려주기

hm = sns.heatmap(
    corr.values,
    cbar= True,
    annot = True, # 히트맵에 숫자 표시
    square = False, 
    fmt = '.2f',
    annot_kws= {'size': 20},
    xticklabels= cols_view,
    yticklabels = cols_view
)

plt.tight_layout()
plt.show()
```
