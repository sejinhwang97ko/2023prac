```python
def load(_dir, _end='.csv'):
    # _dir: 파일의 경로, _end: 파일의 확장자

    if ~(_dir.endswith("/")):
        _dir = _dir+"/"

    if _end[0] != ".":
        _end = "." + _end

    print (_dir, _end)

    # # 파일의 리스트 호출
    file_list = os.listdir(_dir)
    # print(file_list)

    # #비어있는 데이터프레임 생성
    result = pd.DataFrame()
    # # 파일 리스트에서 확장자가 같은 파일들만 결합
    for i in file_list:
        # 파일명의 확장자라는 인자값과 같은 경우
        if i.endswith(_end):
            # 확장자에 따른 read 함수 설정
            if _end == '.csv':
                df = pd.read_csv(_dir+i)
                result = pd.concat([result, df], axis = 0, ignore_index=True)
            elif _end == '.json':
                df = pd.read_json(_dir+i)
                result = pd.concat([result, df], axis = 0, ignore_index=True)
            elif _end in [".xlsx" , "xls"]: #(_end == '.xlsx' or _end == '.xls'): 
                df = pd.read_excel(_dir+i)
                result = pd.concat([result, df], axis = 0, ignore_index=True)
            else:
                return "지원하지 않는 확장자입니다."
    return result
```

## 예제
1. csv 폴더에서 4개의 파일을 로드(tran_1, tran_2, tran_d_1, tran_d_2)
2. tran_1 과 tran_2는 단순한 행 추가 결합(유니언 결합)
3. tran_d_1과 tran_d_2도 유니언 결합
4. 2번, 3번 과정에서 나온 데이터프레임을 특정 조건에 맞춰서 2번 데이터프레임을 기준으로 
열 추가 결합(조인 결합)
5. csv 폴더에 있는 2개의 파일 로드(custom_master, item_master)
6. 4번 과정에서 나온 데이터프레임과 custom_master 조인 결합(inner나 left를 제일 많이 씀)
7. 6번 과정에서 나온 데이터프레임과 item_master 조인 결합
8. 결합이 된 데이터프레임에서 quantity 컬럼과 item_price 컬럼의 데이터를 가지고 total_price라는 파생변수를 생성(total_price = item_price * quantity)

```python
tran_1 = pd.read_csv("../csv/csv/tran_1.csv")
tran_2 = pd.read_csv("../csv/csv/tran_2.csv")
tran_d_1 = pd.read_csv("../csv/csv/tran_d_1.csv")
tran_d_2 = pd.read_csv("../csv/csv/tran_d_2.csv")
```

```python
df1 = pd.concat([tran_1, tran_2], axis = 0, ignore_index=True)
df1
```

```python
df2 = pd.concat([tran_d_1, tran_d_2], axis = 0, ignore_index=True)
df2
```

```python
# 2번, 3번 과정에서 나온 데이터프레임을 특정 조건에 맞춰서 2번 데이터프레임을 기준으로 열 추가 결합(조인 결합)

df3 = pd.merge(df1, df2, on='transaction_id', how='left')
df3
```

```python
# 5. csv 폴더에 있는 2개의 파일 로드(custom_master, item_master)

cus = pd.read_csv("../csv/csv/customer_master.csv")
item = pd.read_csv("../csv/csv/item_master.csv")
```

```python
cus
```

```python
df4 = pd.merge(df3, cus, on='customer_id', how='left')
df4
```

```python
item
```

```python
df5 = pd.merge(df4, item, on='item_id', how='left')
df5
```

```python
# 결합이 된 데이터프레임에서 quantity 컬럼과 item_price 컬럼의 데이터를 가지고
# total_price라는 파생변수를 생성(total_price = item_price * quantity)

df5['total_price'] = df5['item_price'] * df5['quantity']
```

```python
df5
```

```python
# payment_date를 시계열로 #2019-02-01 01:36:57
# item_name 기준으로 total 합계

df5['payment_date'] = pd.to_datetime(df5['payment_date']) #, format="%Y-%m-%d %H:%M:%S"
```

```python
df5.info()
```

```python
df6 = df5.groupby('item_name').sum()
```

```python
df7 = df6.sort_values('total_price', ascending=False)
df7
```
