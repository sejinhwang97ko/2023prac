```python
def halloween(df):
    # datetime 컬럼 하나 만들기
    df['Datetime'] = pd.to_datetime(df['Date'])
    # datetime 컬럼에서 년도, 월 뽑고 컬럼 만들기
    df['year'] = df['Datetime'].dt.year
    df['month'] = df['Datetime'].dt.month
    # 2000년부터 2019년까지 뽑기
    df1 = df.loc[(df['year'] >= 2000) & (df['year'] <= 2019)]
    # datetime 컬럼에서 년월 뽑고 컬럼 만들기
    def change(x):
        return datetime.strftime(x, '%Y-%m')
    df1['년월'] = list(map(change, df1['Datetime']))
    # 첫날만 뽑기
    df2 =df1.loc[df1['년월'] != df1['년월'].shift(1)]
    # 시가
    list_11 = df2.loc[df2['month'] == 11,'Open'].to_list()

    # 마지막날만 뽑기
    df3= df1.loc[df1['년월'] != df1['년월'].shift(-1)]
    # 종가
    list_4 = df3.loc[df3['month'] == 4,'Close'].to_list()

    list_ = []
    for i in range(0, len(list_11)):
        list_.append((list_4[i+1] - list_11[i]) / list_11[i] + 1)

    return list_
```

```python
df = pd.read_csv("../csv/csv/AMZN.csv")
ha = halloween(df)
```

```python
acc_rtn = 1
for i in ha:
    acc_rtn *= i

acc_rtn
```

```python
df = pd.read_csv("../csv/csv/AMZN.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
df.set_index('Date', inplace=True)
```

```python
for i in range(2000,2020,1):
    buy_mon = str(i) + "-11"
    sell_mon = str(i+1) + "-04"
    print(buy_mon, sell_mon)
```

```python
acc_rtn = 1
for i in range(2000,2019,1):
    buy_mon = str(i) + "-11"
    sell_mon = str(i+1) + "-04"
    # 구입가 저장
    buy = df.loc[buy_mon].iloc[0]['Open']
    sell = df.loc[sell_mon].iloc[-1]['Close']

    # 수익율
    rtn = (sell - buy) / buy + 1

    # 누적 수익율
    acc_rtn *= rtn
```

```python
acc_rtn
```

```python
# buyandhold 인 경우
buy = df.loc["2000-11"].iloc[0]['Open']
sell = df.loc["2019-04"].iloc[-1]['Close']

rtn = (sell - buy) / buy + 1
rtn
```

```python
# CAGR(연복리수익율)
# (수익율 ** (1/투자기간)) - 1

# 할로윈 CAGR
CAGR = (acc_rtn ** (1/20)) -1
print(CAGR * 100)
```

```python
# 바이앤홀드 CAGR
CAGR = (rtn ** (1/20)) -1
print(CAGR * 100)
```

```python
def six_month(df, start= 2000, end = 2019, month = 11):
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
        df.set_index('Date', inplace=True)

    acc_rtn=1

    for i in range(start, end):
        start_ = datetime(year=i, month=month, day= 1)
        end_ = start_ + relativedelta(months=5)

        buy_mon = start_.strftime("%Y-%m")
        sell_mon = end_.strftime("%Y-%m")

        buy = df.loc[buy_mon].iloc[0]['Open']
        sell = df.loc[sell_mon].iloc[-1]['Close']

        # 수익율
        rtn = (sell - buy) / buy + 1

        # 누적 수익율
        acc_rtn *= rtn

    return acc_rtn
```

```python
data = pd.read_csv("../csv/csv/AMZN.csv")
```

```python
six_month(data)
```

```python
for i in range(1,13,1):
    print(f"{i}월:", six_month(data, month=i))
```
