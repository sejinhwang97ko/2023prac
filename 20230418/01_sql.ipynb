```python
sql = """
    select
    `Country`,
    `Item Type`,
    AVG(`Units Sold`) as AVG,
    SUM(`Units Sold`) as SUM
    from
    `Sales Records`
    group by
    `Country`,
    `Item Type`
    order by
    `Country`
    ASC
"""

db.func_3(sql)
```

1. emp 데이터에서 파생변수 H/L 생성

```python
emp = pd.read_csv("../csv/csv/emp.csv")
emp.head(5)
```

```python
emp.info()
```

```python
emp['SAL'] = emp['SAL'].astype(int)
```

```python
emp
```

```python
sql = """
    select
    `EMPNO`,
    `ENAME`,
    `SAL`,
    CASE
        WHEN `SAL` >= 2000
        THEN 'High'
        ELSE 'Low'
    END as 'H/L'
    from 
    emp
"""

db.func_3(sql)
```
