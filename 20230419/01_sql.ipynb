# 문제
1. emp  table에서 SAL 1000이상이고 3000미만인 사원의 정보를 출력
2. emp, dept를 left를 이용하여 결합한 뒤 부서의 위치가 'NEW YORK'인 데이터를 출력

```python
import mod_sql
db = mod_sql.Database("localhost", "root" , "0630", "ubion7", 3306)
```

```python
sql = """
    select
    *
    from
    emp
    where
    `SAL` >= 1000 AND `SAL` < 3000
"""

db.func_3(sql)
```

```python
#emp, dept를 left를 이용하여 결합한 뒤 부서의 위치가 'NEW YORK'인 데이터를 출력
sql = """
    select
    *
    from
    emp
    left join
    dept
    on
    emp.deptno = dept.deptno
    where
    dept.loc = 'NEW YORK'
"""

db.func_3(sql)
```

1. emp, dept를 right join 결합
2. 부서번호가 40인 데이터만 출력

```python
sql = """
    select
    *
    from
    emp
    right join
    dept
    on
    emp.deptno = dept.deptno
    where
    emp.deptno = 40
    or
    dept.deptno = 40
"""

db.func_3(sql)
```

```python
sql = """
    select DISTINCT deptno FROM emp
"""

db.func_3(sql)
```

```python
sql = """
    select DISTINCT loc FROM dept
"""

db.func_3(sql)
```
