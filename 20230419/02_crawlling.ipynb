```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
data = requests.get("https://finance.naver.com/marketindex/exchangeList.naver").text
```

```python
soup = bs(data, 'html.parser')
soup
```

```python
# find는 html안에서 해당 태그의 첫번째 정보를 출력
tbody = soup.find('tbody')
```

```python
# find_all()는 해당 태그의 전체 정보를 리스트의 형태로 출력
soup.find_all('td')[0].get_text().replace("\n", "")
```

```python
soup.find_all('td')[0].get_text().replace("\n", "").replace("\t", "")
# soup.find_all('td')[0].get_text().replace("\n", "").strip("\t")
```

```python
data_list = soup.find_all('td')
```

```python
# data_list를 이용하여 2차원 배열로 새로운 리스트를 생성
_list = list()
for i in data_list:
    x = i.get_text()
    x = x.replace("\n", "")
    x = x.replace("\t", "")
    _list.append(x)
_list
```

```python
result = []
for i in range(len(_list)//7):
    x = []
    for j in range(i*7, (i+1)*7, 1):
        x.append(_list[j])
    result.append(x)
result
```

```python
result = list()
for i in range(len(_list)//7):
    j = i*7
    result.append(_list[j:j+7])

result
```
