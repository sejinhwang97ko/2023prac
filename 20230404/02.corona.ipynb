```python
corona = pd.read_csv("../csv/csv/corona.csv")
corona
```

```python
corona = corona.drop('Unnamed: 0', axis=1)
```

```python
corona.head(1)
```

```python
corona.columns = ['등록일시', '누적사망자', '누적확진자', '게시글번호', '기준일', '기준시간', '수정일시',
'누적의심자', '누적확진률']
```

```python
corona.sort_values("등록일시", inplace=True)
```

```python
corona
```

```python
#corona.reset_index(drop=True, inplace=True)
```

```python
corona.reset_index()
```

```python
corona.reset_index().drop("index", axis=1, inplace=True)
corona
```

```python
# 데이터간의 차이를 출력하는 함수
# diff(n)
# n의 기본값은 1 -> 1 전 인덱스 데이터를 선택

corona['확진자'] = corona['누적확진자'].diff(1)
```

```python
# shift()
# 이동한 값을 출력

corona2 = corona.head(5)
```

```python
corona2['누적사망자']
```

```python
corona2['누적사망자'].shift(1)
```

```python
corona['일일사망자'] = corona['누적사망자'] - corona['누적사망자'].shift(1)
corona.head(2)
```
