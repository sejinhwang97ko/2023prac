```python
import pandas as pd
```

```python
df = pd.read_csv("https://bit.ly/ds-house-price")
df
```

```python
# 컬럼의 이름을 변경
df = df.rename(columns={'분양가격(㎡)' :'분양가격'})
df
```

```python
df.info()
```

1. 분양가격 컬럼의 데이터의 형태를 int 형태로 변경
2. 규모구분 컬럼의 전용면적 단어를 삭제
3. 지역, 규모구분 별 분양가격의 평균가

```python
df.loc[df['분양가격'].isnull(), '분양가격'] = 0
```

```python
df
```

```python
df['분양가격'] = df['분양가격'].replace('  ', 0)
```

```python
df['분양가격'] = df['분양가격'].str.replace(',', '')
```

```python
df.loc[df['분양가격'].isnull(), '분양가격'] = 0
```

```python
df['분양가격'] = df['분양가격'].str.replace('-', '')
```

```python
df.loc[df['분양가격'].isnull(), '분양가격'] = 0
```

```python
df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df['분양가격'] = df['분양가격'].astype('int')
```

```python
df['분양가격'].isnull().sum()
```

```python
df.info()
```

```python
df['규모구분'] = df['규모구분'].str.replace('전용면적', '')
```

```python
df['규모구분'].str.contains('전용면적')
```

```python
df[['지역명','규모구분','분양가격']].groupby(['지역명','규모구분']).mean()
```
